# 05. Model Evaluation & Hyperparameter Tuning
This notebook evaluates our XGBoost model on the untouched test set and visually compares it to our Tuned model.

```python
import pandas as pd
import joblib
from IPython.display import Image, display
import sys
sys.path.append('..')
from src.models.evaluate_model import evaluate_model_metrics
```

## 1. Load Data and Models

```python
X_test = pd.read_parquet('../data/interim/X_test.parquet')
y_test = pd.read_parquet('../data/interim/y_test.parquet')['Is Laundering']

base_xgb = joblib.load('../models/advanced_xgboost.joblib')
try:
    tuned_xgb = joblib.load('../models/tuned_xgboost.joblib')
except:
    print('Tuned model not yet generated. Run run_evaluation.py first.')
    tuned_xgb = None
```

## 2. Compare PR-AUC and Classification Reports

```python
probs_base, preds_base, auc_base = evaluate_model_metrics(base_xgb, X_test, y_test, 'Base XGBoost')
if tuned_xgb:
    probs_tuned, preds_tuned, auc_tuned = evaluate_model_metrics(tuned_xgb, X_test, y_test, 'Tuned XGBoost')
```

## 3. Visualizations
Visualizing the comparative Precision-Recall curves and Confusion Matrices.

```python
print('--- Base Confusion Matrix ---')
display(Image(filename='../reports/figures/base_xgboost_cm.png'))
print('\n--- Tuned Confusion Matrix ---')
display(Image(filename='../reports/figures/tuned_xgboost_cm.png'))
print('\n--- PR Curve Comparison ---')
display(Image(filename='../reports/figures/pr_curve_comparison.png'))
```
