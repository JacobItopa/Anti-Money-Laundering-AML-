# 03. Model Selection & Training
This notebook demonstrates how we split our massive dataset and train our Baseline and Advanced algorithms.
We use `XGBoost` configured with `scale_pos_weight` to natively handle the 1:980 class imbalance.

```python
import pandas as pd
import sys
sys.path.append('..')

from src.models.train_model import split_data, train_baseline_model, train_xgboost_model
```

## 1. Load Processed Data

```python
df = pd.read_parquet('../data/processed/processed_transactions.parquet')
print(f"Data loaded successfully. Shape: {df.shape}")
```

We perform a **Stratified Split** (80/20) to ensure the extremely rare illicit transactions are proportionally distributed.

```python
X_train, X_test, y_train, y_test = split_data(df, target_col='Is Laundering')
```

## 3. Establish a Baseline Model
Using Logistic Regression with `class_weight='balanced'` to set our performance floor.

```python
# Training the baseline
baseline_model = train_baseline_model(X_train, y_train)
```

## 4. Train Advanced Model (XGBoost)
Using XGBoost with `tree_method='hist'` for blazing fast training on 5 million rows, and dynamically calculating `scale_pos_weight`.

```python
# Training XGBoost
xgb_model = train_xgboost_model(X_train, y_train)
```
